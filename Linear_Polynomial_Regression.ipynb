# Evaluation Metrics

Evaluation Metrics are like tools we use to measure how well a ML Model is doing its job

Types:
* Mean Squared Error (MSE)
  * MSE = ∑(actual - predicted)^2 / n
* R-Squared Value (R2)
* Mean Absolute Error (MAE)
  * MAE = ∑|actual - predicted| / n
* Root Mean Squared Error (RMSE)
  * RMSE = √(∑(actual - predicted)^2 / n)
* Mean Absolute Percentage Error (MAPE)
  * MAPE = (∑|actual - predicted| / actual) * (100 / n)

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
```

```python
mean_squared_error(y_test, predictions)
```

```python
r2_score(y_test, predictions)
```

```python
mean_absolute_error(y_test, predictions)
```

```python
mean_absolute_percentage_error(y_test, predictions)
```

```python
import numpy as np

np.sqrt(mean_squared_error(y_test, predictions))
```

# Polynomial

It is a type of Regression analysis used in ML and Statistics to model relationship between variables. In simple Linear Regression, we use a straight line to fit the data points, but in Polynomial we use a polynomial Equation of a higher degree to capture more patterns in data

Degree:

  * 0 = Constant Function
  * 1 = Linear Function (Straight Line)
  * 2 = Quadratic Function (U - Shaped Curve)
  * 3 = Cubic Function (S - Shaped Curve)

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False)
```

```python
x_poly = poly_features.fit_transform(x_train)

X_test = poly_features.fit_transform(x_test)
```

```python
model = LinearRegression().fit(x_poly, y_train)
```

```python
final = model.predict(X_test)
```

```python
mean_squared_error(y_test, final)
```

```python
r2_score(y_test, final)
```

```python
mean_absolute_error(y_test, final)
```

```python
mean_absolute_percentage_error(y_test, final)
```

```python
np.sqrt(mean_squared_error(y_test, final))
```
